```python
import pandas as pd
# The "Strategy" or "Check" definition (simplified)
class Check:
    def __init__(self, name, validation_func, *args, **kwargs):
        self.name = name
        self.func = validation_func
        self.args = args
        self.kwargs = kwargs

    def __call__(self, series: pd.Series) -> pd.Index:
        # Executes the specific validation logic
        return self.func(series, *self.args, **self.kwargs)

# The "Composite" or "Schema" orchestrator
class DataFrameValidator:
    def __init__(self):
        # A dictionary where keys are column names, values are lists of checks
        self.schema = {}

    def add_column_checks(self, column_name: str, checks: list[Check]):
        self.schema[column_name] = checks

    def validate(self, df: pd.DataFrame) -> dict:
        results = {}
        for col_name, checks in self.schema.items():
            for check in checks:
                errors = check(df[col_name])
                if not errors.empty:
                    results[f"{col_name}_{check.name}"] = errors
        return results
```
