```python
import polars as pl
import numpy as np

# Sample DataFrame with null and NaN values
df = pl.DataFrame({
    "A": [1, None, 3],
    "B": [4.0, 5.0, np.nan], # NaN is specific to float columns
    "C": ["x", "y", None]
})

df
```

```python
import polars.selectors as cs

def is_empty(df: pl.DataFrame) -> bool:
    return df.with_columns(
        pl.all().is_null()
    ).with_columns(
        cs.float().is_nan()
    )

is_empty(df)
```

```python
df1 = df.clone()
for col in df1.columns:
    col_type = df1[col].dtype
    print(col, col_type)
    if col_type in ["float64", "float32"]:
        mask = df1.with_columns(
            cs.float().is_nan().alias(f"{col}_is_nan")
        )
        print(mask)
    else:
        mask = df1.with_columns(
            cs.all().is_null().alias(f"{col}_is_null")
        )
        print(mask)
df1
```
