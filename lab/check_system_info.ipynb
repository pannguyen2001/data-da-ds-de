## Read file, folder in a folder

```python
import os
import sys
from os import walk

my_path: str = r"/home/user/prj1mrdp/logs"
f = []
for (dir_path, dir_names, file_names) in walk(my_path):
    print(f"{'':=^30}")
    print(f"{dir_path = }, {dir_names = }, {file_names = }")
    f.extend(file_names)

file_paths: list = []
for file_name in f:
    file_path: str = os.path.join(my_path, file_name)
    file_paths.append(file_path)
file_paths
```

## system info

```python
import psutil

print(f"\nCPU Cores: {psutil.cpu_count(logical=False)} physical, {psutil.cpu_count(logical=True)} logical")
print(f"Total RAM: {round(psutil.virtual_memory().total / (1024**3), 2)} GB")
print(f"Available RAM: {round(psutil.virtual_memory().available / (1024**3), 2)} GB")
```
