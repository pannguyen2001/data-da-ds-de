## way 1: ExitStack

```python
from contextlib import ExitStack

with ExitStack() as stack:
    files = [stack.enter_context(open(i, "r")) for i in files]
    for rows in zip(*files):
        # rows is now a tuple containing one row from each file
        print(rows)
```

## file excel/csv with pandas

```python
# Create sample data test
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "col_1": np.ones(1_000, dtype=int),
    "col_2": np.ones(1_000, dtype=int)
})

with pd.ExcelWriter("/home/user/prj1mrdp/data/test/read_multiple_files_01.xlsx") as writer:
    df.to_excel(writer, sheet_name="Sheet1", index=False)
    df.to_excel(writer, sheet_name='Sheet2', index=False)

with pd.ExcelWriter("/home/user/prj1mrdp/data/test/read_multiple_files_02.xlsx") as writer:
    df.to_excel(writer, sheet_name="Sheet1", index=False)
    df.to_excel(writer, sheet_name='Sheet2', index=False)
```

```python
from os import walk

folder_path: str = "/home/user/prj1mrdp/data/test"
file_path: list[str] = []

for _, _, file_name in walk(folder_path):
    file_path.extend(file_name)

file_excel_path: list[str] = []
for file_excel in file_path:
    file_excel_path.append(f"{folder_path}/{file_excel}")
file_excel_path
```

```python
import pandas as pd
import numpy as np

df_dict_storage: dict[pd.DataFrame] = {}

for file_excel in file_excel_path:
    print(file_excel)
    xlsx = pd.ExcelFile(file_excel, engine="calamine")
    for sheet_name in xlsx.sheet_names:
        print(sheet_name)
        df = xlsx.parse(sheet_name)
        df_dict_storage.update({
            f"{file_excel.split('.')[-2].split('/')[-1]}_{sheet_name}": df
        })
```
